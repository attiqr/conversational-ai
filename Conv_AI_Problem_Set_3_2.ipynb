### Importing Packages

```python
from wordcloud import WordCloud, STOPWORDS,  ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

### Mounting Google Drive

### Loading Text File

```python
with open("/content/drive/MyDrive/JaneEyre.txt", "r") as file:
  text = file.read()
```

### Creating WordCloud

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    stopwords=set(STOPWORDS),
    min_font_size=10,
  ).generate(text)
```

### Showing Wordcloud

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

### Customization

```python
! wget "https://amueller.github.io/word_cloud/_images/sphx_glr_masked_002.png"
```

```python
# Load Mask Image
alice_mask = np.array(Image.open("sphx_glr_masked_002.png"))
```

```python
stopwords = set(STOPWORDS) | {"said", "will"}


wc = WordCloud(background_color="lightblue", max_words=5000, mask=alice_mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue', colormap='viridis')

# Generate word cloud
wc.generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
